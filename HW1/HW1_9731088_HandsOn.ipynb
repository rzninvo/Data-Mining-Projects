```python
import pandas as pd
df = pd.read_csv("IMDB-Movie-Data.csv")
df.head(10)
```

```python
pd.isna(df)
```

```python
df.isna().sum()
```

```python
df.isna().query('`Revenue (Millions)`== True and Metascore == True')
```

```python
df = df.dropna(subset=['Revenue (Millions)', 'Metascore'], how='all')
df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
col_mean = df.loc[:, 'Revenue (Millions)'].mean()
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(col_mean)
df
```

```python
df.isna().sum()
```

```python
df = df.sort_values(by=['Rating'], ascending= False)
df.reset_index(inplace=True, drop=True)
df
```

```python
df.drop(columns={'Rank'}, inplace=True)
df.insert(loc=0, column='Rank', value=np.arange(1, df.index.stop+1))
df
```

```python
from sklearn.preprocessing import StandardScaler
col_mean = df.loc[:, 'Metascore'].mean()
df['Metascore'] = df['Metascore'].fillna(col_mean)
df.isna().sum()
```

```python
scaler = StandardScaler()
scaler.fit(df['Metascore'].to_numpy().reshape(-1, 1))
print('Mean=' + str(df['Metascore'].mean()) + ' Variance= ' + str(df['Metascore'].var()))
normalized_col = pd.Series(scaler.transform(df['Metascore'].to_numpy().reshape(-1, 1)).reshape(-1))
print('------Normalized Metascore Column--------')
print(normalized_col)
print('Mean=' + str(normalized_col.mean()) + ' Variance= ' + str(normalized_col.var()))
df['Metascore'] = normalized_col
df
```
