<font color='Green'> Question: Calculate the elbow method for the number of clusters from 1 to 10 and plot the cost function for each cluster. Consider <code>random_state=0</code> for each step. </font>

<h2>Complex Clustering</h2>

Clustering is not always like this. Also, k-means is not the only clustering algorithm. Below, different distributions of different data have been plotted with the k-means method.
You can see in which case this algorithm works well.

All of these codes are using the <code>sklearn</code> documentation. You can see these documents <a href="https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_assumptions.html">here</a>.

You can also see other algorithms in the <code>sklearn</code> documents.
</div>

```python
sns.set_style('white')

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Incorrect number of clusters
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")

# Anisotropicly distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.subplot(222)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Anisotropicly Distributed Blobs")

# Different variance
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)

plt.subplot(223)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Unequal Variance")

# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_filtered)

plt.subplot(224)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs")

plt.show()
```

<h1>DBSCAN Algorithm</h1>

The DBSCAN algorithm is another algorithm used for clustering data, which can cluster data with different shapes and detect noise and anomalies in the data.

In this section, we want to cluster the two datasets mentioned in the "More Complex Clustering" section again using the DBSCAN algorithm. Consider the datasets X_aniso and X_varied.

To run the DBSCAN algorithm, there is no need to determine the number of clusters, and you need to determine the two main parameters of this algorithm, namely epsilon and the minimum number of points in each cluster, in such a way that your output is more accurate.


</div>





<div dir>
<h1>Finding Epsilon</h1>
In order to find the best epsilon for a given dataset, in addition to trying different values, the KNN distance method (calculating the average distance between each point and its k nearest neighbors) is also used to automatically determine the optimal value of epsilon.

Here, k is equal to the minimum number of points in a cluster, which is equivalent to MinPts. Unlike the epsilon parameter, there is no automatic method for determining MinPts. However, there are some general rules based on the characteristics of the dataset that can help us find a suitable range of MinPts values.

<font color='Green'>Question: Explain the reason for using KNN to find the optimal epsilon. Then, by plotting the k-distance graph and following the elbow pattern, find the optimal value of epsilon.

<font color='Green'>To plot the k-distance graph, you can use the NearestNeighbors module in the Scikit-learn library, and to determine the value of epsilon, you can use the kneed library.
</font>

<font color='white'>Answer: We use the KNN algorithm to have a suitable idea of the average distance of each point from its K nearest neighbors so that we can intelligently choose the value of epsilon.</font>

</div>




```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import v_measure_score
from kneed import KneeLocator
import pandas as pd 
```

```python
def DBSCAN_Kneepoint(dataset):

  # Calculate the average distance between each point in the data set and its 10 nearest neighbors
  neighbors = NearestNeighbors(n_neighbors=10)
  neighbors_fit = neighbors.fit(dataset)
  print(f'({neighbors_fit})')
  distances, indices = neighbors_fit.kneighbors(dataset)
  print(f'({distances}, {distances.shape})')

  # Sort distance values by ascending value and plot
  distances = np.sort(np.average(distances, axis=1))
  plt.title("k-distance graph")
  plt.plot(distances)

  i = np.arange(len(distances))
  kl = KneeLocator(i, distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')
  print(f'KneeLocator = {kl}')
  fig = plt.figure()
  kl.plot_knee()
  plt.xlabel("Points")
  plt.ylabel("Distance")

  print("eps: ", distances[kl.knee])
  return distances[kl.knee]
```

```python
epsilon_ansio = DBSCAN_Kneepoint(X_aniso)
```

```python
epsilon_varied = DBSCAN_Kneepoint(X_varied)
```

<h1>Determining MinPts</h1>
In the previous step, we determined the optimal value of eps. However, this may not necessarily result in the best clustering.

<font color='Green'>Question: In this step, considering the obtained eps value and testing different values for the MinPts parameter, plot the best clustering using an appropriate color scheme. </font>

</div>




```python
# your code here
def MinPts_DBSCAN(eps, dataset):
  v_measure_array = []
  fig = plt.figure(figsize=(20, 20))
  fig.subplots_adjust(hspace=.5, wspace=.2)
  i = 1
  fig.suptitle('Various MinPts Plots')
  for j in range(3, 13):
      db = DBSCAN(eps=eps, min_samples=j).fit(dataset)
      core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
      core_samples_mask[db.core_sample_indices_] = True
      labels = db.labels_

      v_measure_array.append(v_measure_score(y, labels))
      
      ax = fig.add_subplot(4, 5, i)
      ax.set_title("minPts = {}".format(round(j)), fontsize=10, ha="center")
      sns.scatterplot(x = dataset[:,0], y = dataset[:,1], hue=["C: {}".format(x) for x in labels])
      ax.legend(loc = "upper right")
      ax_1 = fig.add_subplot(4, 5, 10 + i)
      ax_1.set_title("minPts = {}".format(round(j)), fontsize=10, ha="center")
      sns.scatterplot(x = dataset[:,0], y = dataset[:,1], hue=["Core: {}".format(x) for x in core_samples_mask])

      i += 1

  print("v-measures: ", v_measure_array)
```

```python
MinPts_DBSCAN(epsilon_ansio, X_aniso)
```

<div dir>
<font color='white'> As can be seen from the v-measure values for different minPts values, minPts=3 had the highest v-measure value, so we use it as the final minPts value. </font>
</div>

```python
def DBSCAN_Algorithm(eps, minpts, dataset):
  db = DBSCAN(eps=eps, min_samples=minpts).fit(dataset)

  plt.scatter(dataset[:, 0], dataset[:, 1], c=db.labels_, label=y)
  plt.title('Best Hyperparamters DBScan')

  labels = db.labels_
  num_clusters = len(set(labels))-(1 if -1 in labels else 0)

  print('number of clusters: ', num_clusters)
  print('v_measure: ', v_measure_score(y, labels))
```

```python
DBSCAN_Algorithm(epsilon_ansio, 3, X_aniso)
```

```python
MinPts_DBSCAN(epsilon_varied, X_varied)
```

<font color='white'> As we can see from the v-measure values for different minPts, minPts=3 had the highest v-measure value, so we use it as the final value for minPts. </font>
</div>
